```python
# Add Dependencies 

import random as r
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import requests
import traceback
from citipy import citipy
```

```python
## Generate a new set of 1,500 random latitudes and longitudes.
# combine random.randint(-90, 90) and random.random() to generate a floating-point decimal between –90 and 90, to generate a random latitude.

# Assign the variable x to 1.
x =1
# Initialize an empty list, latitudes.
latitudes = []
# Create a while loop where we generate a random latitude and add it to the list.
while x < 1501:
    random_lat = r.randint(-90, 90) + r.random()
    latitudes.append(random_lat)
    # After the random latitude is added to the list we add one to the variable “x”.
    x += 1
    
# The while loop condition is checked again and will continue to run as long as x is less than 11.

latitudes
```

```python
# combine random.randint(-180, 180) and random.random() to generate a floating-point decimal between –180 and 180, to generate a random longitude.

# Assign the variable x to 1.
x =1
# Initialize an empty list, longitudes.
longitudes = []
# Create a while loop where we generate a random latitude and add it to the list.
while x < 1501:
    random_lng = r.randint(-90, 90) + r.random()
    longitudes.append(random_lng)
    # After the random latitude is added to the list we add one to the variable “x”.
    x += 1
    
# The while loop condition is checked again and will continue to run as long as x is less than 11.

longitudes
```

```python
# Merge into list
lats_lngs = zip(latitudes, longitudes)
lats_lngs
```

```python
## Get the nearest city using the citipy module.

# Add the lats_lngs to a list.
coordinates = list(lats_lngs)
```

```python
# Create tuple to display the lats and lngs combinations.
for coordinate in coordinates:
    print(citipy.nearest_city(coordinate[0], coordinate[1]).city_name,
          citipy.nearest_city(coordinate[0], coordinate[1]).country_code)
```

```python
# Create a list for holding the cities.
cities = []
# Identify the nearest city for each latitude and longitude combination.
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    if city not in cities:
        cities.append(city)
    
print(cities)
```

```python
# Check legnth
print(len(cities))
```

```python
## Perform an API call with the OpenWeatherMap.

# Import the API key.
from config import weather_api_key
# Import the datetime module.
from datetime import datetime
```

```python
# Starting URL for Weather Map API.
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
print(url)
```

```python
## Retrieve the following information from the API call. --> Lat and long, max temp, % humidity, % cloudiness, wind speed, weather description (e.g., clouds, fog, light rain, clear sky)

# Create an empty list to hold the weather data.
city_data = []
# Loop through all the cities in our list.
for city in cities:
    # Create endpoint URL with each city.
    city_url = url + "&q=" + city
    # conditional for 404 code
    city_weather = requests.get(city_url).json()
    code = city_weather["cod"]
    
    if code == 200:
        try: # Rain only.
            city_weather = requests.get(city_url).json()
            city_lat = city_weather["coord"]["lat"]
            city_lng = city_weather["coord"]["lon"]
            city_max_temp = city_weather["main"]["temp_max"]
            city_humidity = city_weather["main"]["humidity"]
            city_clouds = city_weather["clouds"]["all"]
            city_wind = city_weather["wind"]["speed"]
            city_country = city_weather["sys"]["country"]
            city_desc = city_weather["weather"][0]["description"]
            city_rain = city_weather["rain"]["3h"]
            city_date = datetime.utcfromtimestamp(city_weather["dt"]).strftime('%Y-%m-%d %H:%M:%S')
            
            # Rain & no snow.
            city_data.append({"City": city.title(),
                              "Country": city_country,
                              "Date": city_date,
                              "Lat": city_lat,
                              "Lng": city_lng,
                              "Max Temp": city_max_temp,
                              "Humidity": city_humidity,
                              "Cloudiness": city_clouds,
                              "Wind Speed": city_wind,
                              "Current Description": city_desc,
                              "Rain Inches (last 3 hrs)": city_rain,
                              "Snow Inches (last 3 hrs)": 0})
        except: # If no Rain...
            try: # With Snow.
                city_weather = requests.get(city_url).json()
                city_lat = city_weather["coord"]["lat"]
                city_lng = city_weather["coord"]["lon"]
                city_max_temp = city_weather["main"]["temp_max"]
                city_humidity = city_weather["main"]["humidity"]
                city_clouds = city_weather["clouds"]["all"]
                city_wind = city_weather["wind"]["speed"]
                city_country = city_weather["sys"]["country"]
                city_desc = city_weather["weather"][0]["description"]
                city_snow = city_weather["snow"]["3h"]
                city_date = datetime.utcfromtimestamp(city_weather["dt"]).strftime('%Y-%m-%d %H:%M:%S')
                
                # With Snow & no rain.
                city_data.append({"City": city.title(),
                                  "Country": city_country,
                                  "Date": city_date,
                                  "Lat": city_lat,
                                  "Lng": city_lng,
                                  "Max Temp": city_max_temp,
                                  "Humidity": city_humidity,
                                  "Cloudiness": city_clouds,
                                  "Wind Speed": city_wind,
                                  "Current Description": city_desc,
                                  "Rain Inches (last 3 hrs)": 0,
                                  "Snow Inches (last 3 hrs)": city_snow})
                
            except: # If no rain or snow...
                # change to snow and rain as 0.
                city_weather = requests.get(city_url).json()
                city_lat = city_weather["coord"]["lat"]
                city_lng = city_weather["coord"]["lon"]
                city_max_temp = city_weather["main"]["temp_max"]
                city_humidity = city_weather["main"]["humidity"]
                city_clouds = city_weather["clouds"]["all"]
                city_wind = city_weather["wind"]["speed"]
                city_country = city_weather["sys"]["country"]
                city_desc = city_weather["weather"][0]["description"]
                city_date = datetime.utcfromtimestamp(city_weather["dt"]).strftime('%Y-%m-%d %H:%M:%S')
                
                city_data.append({"City": city.title(),
                                  "Country": city_country,
                                  "Date": city_date,
                                  "Lat": city_lat,
                                  "Lng": city_lng,
                                  "Max Temp": city_max_temp,
                                  "Humidity": city_humidity,
                                  "Cloudiness": city_clouds,
                                  "Wind Speed": city_wind,
                                  "Current Description": city_desc,
                                  "Rain Inches (last 3 hrs)": 0,
                                  "Snow Inches (last 3 hrs)": 0})
                    
    else: #if the code is 404
        pass  
```

```python
## Add data to new DataFrame.

city_data_df = pd.DataFrame(city_data)
city_data_df.head()
```

```python
# Save the new dataframe as a CSV file to be used for Part 2.

city_data_df.to_csv('data\WeatherPy_challenge.csv', index=False)
```

```python
# Answer "How many cities have recorded rainfall and snow?" using pandas.

cities_rain = (city_data_df['Rain Inches (last 3 hrs)'] != 0).sum()
cities_snow = (city_data_df['Snow Inches (last 3 hrs)'] != 0).sum()

print(f"There are {cities_rain} cities that have recorded rainfall, and {cities_snow} cities with recorded snow.")
```
